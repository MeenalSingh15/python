#### Pandas Series

A Pandas series is a one-dimensional array which can hold various data types. It is similar to a Python list and a NumPy array

```python
# list or numpy array 1-d is equivalent to series
import random
l=[random.randint(30,69)for i in range(5)]
l
print(l,type(l))
```

```python
import pandas as pd
weights=pd.Series(l)
print(weights,type(weights))
```

You can use the size keyword to find the number of items in a Pandas series.


You can use the shape keyword to find the number of rows and columns in a Pandas series.



```python
print(weights.shape)
print(weights.size)
```

```python
roll=pd.Series(list(range(5)))
```

**Concat:**


Combines multiple Pandas objects (Series or DataFrame) along a particular axis, creating a new, unified object.

It's a powerful tool for assembling larger datasets from smaller pieces.

```python
data1=pd.concat([roll,weights],axis=1)#columns are joined along axis rows you are operating horizontally,
data2=pd.concat([roll,weights],axis=0)
print(data1,'\n \n')
print(data2)
```

```python
#  now combination of series results in a data frame
print(type(data1),type(data2))
```

**set_index():**

The Pandas Dataframe.set_index() method is used to assign a list, series, or another data frame as the index of a given data frame.

Syntax: DataFrame.set_index(keys, drop=True, inplace=False)

This operation does not modify the original DataFrame unless you use the inplace=True parameter.

```python
print(df.set_index('Grade'))
print(df,'See Df does not has grade as index because inplace was false in code')
```

```python
df.set_index("Grade",inplace=True)
df
```

```python
df.index
```

```python
df.columns
```

**df.reset_index():**

Moves the current index into a column and creates a new numerical index.

The reset_index() method allows you reset the index back to the default 0, 1, 2 etc indexes.

By default this method will keep the "old" idexes in a column named "index", to avoid this, use the drop parameter.

```python
di={"name":['Meenal',"Justin","Lucifer"],"Age":[25,27,30],"city":["NewYork","Goa","LosAngeles"],"Grade":["A",'B','C']}
daf=pd.DataFrame(di)
daf
```

```python
daf.set_index('Grade',inplace=True)
print(daf)
```

```python
daf.reset_index()
```
