**Sets**
Sets are used to store multiple items in a single variable.

Set is one of 4 built-in data types in Python used to store collections of data, the other 3 are List, Tuple, and Dictionary, all with different qualities and usage.

A set is a collection which is unordered, unchangeable*, and unindexed.

* Note: Set items are unchangeable, but you can remove items and add new items.

Set items are unordered, unchangeable, and do not allow duplicate values.


 *Unordered* means that the items in a set do not have a defined order.

Set items can appear in a different order every time you use them, and cannot be referred to by index or key.


Set items are *unchangeable*, meaning that we cannot change the items after the set has been created.

*Once a set is created, you cannot change its items, but you can remove items and add new items.*

 **Duplicates Not Allowed**
Sets cannot have two items with the same value.

```python
fruits = {"apple", "banana", "cherry", "apple","cherry"}

print(fruits)# Duplicates are  omitted
```

```python
unordered={"i","am","unordered","can","you","see?",}# result is not in order it was created there is no concept of index here
print(unordered)
```

**Set methods**

```python
# You cannot access items in a set by referring to an index or a key.

# But you can loop through the set items using a for loop, 
 # or ask if a specified value is present in a set, by using the in keyword.
for x in fruits:
  print(x)

```

```python
# Check if "banana" is present in the set:


print("banana" in fruits)
```

Once a set is created, you cannot change its items, but you can add new items.

To add one item to a set use the add() method.

```python
fruits.add("orange")

print(fruits)
```

To add items from another set into the current set, use the update() method.The object in the update() method does not have to be a set, it can be any iterable object (tuples, lists, dictionaries etc.).

```python
myfav={"mango","guava","litchi"}
fruits.update(myfav)# updates fruits set with contents of myfav set
print(fruits)
```

```python
lis=[1,2,3,4,5,5]
fruits.update(lis)# updates fruits set with contents of lis list
print(fruits)
```

**union** method
Basically combination of all elements present in both the sets .

```python
s1={1,2,3,4}
s2={3,4,5,6}# 3,4 common in both
print(s1.union(s2))
print(s1,s2)
```

```python
s1.update(s2)#update method changes values in s1
print(s1,s2)
```

**intersection()
intersection_update()**
gives only values common to both sets and not all values

```python
city1={"Lucknow","Paris","Madrid","Seoul","Denver"}
city2={"Lucknow","Delhi","Bhopal","Goa"}
city3=city1.intersection(city2)#gives only values common to both sets and not all values
print(city3)
print(city1,city2,city3)
```

```python
city1.intersection_update(city2)# updates intersection value of city1 and  city2 to city1
print(city1,city2)
```

**The symmetric_difference()**
method gives you the symmetric difference between two sets. It gives you a set of elements that exist in either of the sets but not in both.





```python
x={1,2,3,4,5,6,7,"Meenal","Singh"}
y={1,3,4,5,88,99,00,100}
z=x.symmetric_difference(y)# gives values not common in both sets
print(z)
print(x,y)
```

```python
x.symmetric_difference_update(y)
print(x,y)# value of x is updated with value of symmetric difference of x and y
```

**The difference()** method returns a set that contains the difference between two sets.

Meaning: The returned set contains items that exist only in the first set, and not in both sets.

```python
a={1,2,3,4,5,6,7}
b={5,6,78,7,8,9}
print(a.difference(b))# a-b basically all common elements of a and b are removed from a
print(a,b)
```

```python
a.difference_update(b)# a gets updated with a-b value
print(a,b)
```

*isdisjoint()* checks if two sets are disjoint and have no common value

```python
name={"Ram","Rohan","Rahul","Yash"}
num={1,2,3,4,5}
print(name.isdisjoint(num))# no common values so they are disjoint
```

**The issuperset()** is a method in Python that helps you find out if a set is a superset of another set. A superset is usually a set that contains one or more subsets.

```python
A = {12, 14, 18, 22, 24}
B = {12, 14, 18}
C = {22, 24}

print (A. issuperset(B))          # returns True because A has all elements of B
print (C. issuperset(A))          # returns False because C does not have all elements of A
print (A. issuperset(C))          # returns True because A has all elements of C
```

Similar to the issuperset() in Python,**issubset()** is a method that helps you compare two relations between two sets. The method returns True if the set is a subset of the argument, and returns False if it is not a subset.

```python
h={"h",'i','m','y','n','a','m','e','is','meenal'}
greet={"h",'i'}
name={'m','y','n','a','m','e','is','meenal'}

print(h.issubset(greet))# returns false because greet does not have all elements of h
print(greet.issubset(h))# returns true all elements of greet in h
print(name.issubset(h))# returns true all elements of name in h
print(h.issubset(name))# returns false all elements of h not in name
```

Remove Item
To remove an item in a set, use the remove(), or the discard() method.
The discard() method removes the specified item from the set. This method is different from the remove() method, because the remove() method will raise an error if the specified item does not exist, and the discard() method will not.

```python
my={1,2,3,4,5,6,7,8}
my.remove(1)# removes 1 from set my
print (my)
```

```python
my.remove(11)# will raise an error because 11 does not exist in my
```

```python
my.discard(11)# will not raise any error
```

```python
my.discard(5)# will discard 5 from set my
my
```

```python
mmy={'hi','hello','hola',1,2,3,4,5}
poped=mmy.pop()#removes any random element from the set and returns the removed element
print(mmy,poped)
```
